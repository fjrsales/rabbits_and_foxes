```python
import random
import time
from collections import defaultdict
from IPython.display import Image, display

# counter.py
class Counter:
    """
    Fornece um contador para um participante na simula√ß√£o.
    Inclui uma string identificadora e um contador de quantos
    participantes deste tipo existem atualmente na simula√ß√£o.
    """

    def __init__(self, name):
        """
        Fornece um nome para um dos tipos de simula√ß√£o.
        :param name: Um nome, ex: "Fox"
        """
        self.name = name
        self.count = 0

    def get_name(self):
        """Retorna a descri√ß√£o curta deste tipo."""
        return self.name

    def get_count(self):
        """Retorna o contador atual para este tipo."""
        return self.count

    def increment(self):
        """Incrementa o contador atual em um."""
        self.count += 1

    def reset(self):
        """Redefine o contador atual para zero."""
        self.count = 0

# location.py
class Location:
    """Representa uma localiza√ß√£o em uma grade retangular."""

    def __init__(self, row, col):
        self.row = row
        self.col = col

    def __eq__(self, other):
        if isinstance(other, Location):
            return self.row == other.row and self.col == other.col
        return False

    def __hash__(self):
        return hash((self.row, self.col))

    def __str__(self):
        return f"Location({self.row}, {self.col})"

    def __repr__(self):
        return self.__str__()


# randomizer.py
class Randomizer:
    """
    Fornece controle sobre a aleatoriedade da simula√ß√£o. Usando o
    randomizador compartilhado com semente fixa, execu√ß√µes repetidas
    ter√£o exatamente o mesmo comportamento (√∫til para testes).
    """

    # A semente padr√£o para controle da aleatoriedade
    SEED = 1111
    # Um objeto Random compartilhado, se necess√°rio
    _rand = random.Random(SEED)
    # Determina se um gerador aleat√≥rio compartilhado deve ser fornecido
    USE_SHARED = True

    @staticmethod
    def get_random():
        """Fornece um gerador aleat√≥rio."""
        if Randomizer.USE_SHARED:
            return Randomizer._rand
        else:
            return random.Random()

    @staticmethod
    def reset():
        """
        Redefine a aleatoriedade.
        N√£o ter√° efeito se a aleatoriedade n√£o for atrav√©s de um
        gerador Random compartilhado.
        """
        if Randomizer.USE_SHARED:
            Randomizer._rand.seed(Randomizer.SEED)


# rabbit.py
class Rabbit:
    """
    Um modelo simples de um coelho.
    Coelhos envelhecem, se movem, procriam e morrem.
    """

    # Caracter√≠sticas compartilhadas por todos os coelhos (vari√°veis de classe)
    BREEDING_AGE = 5  # A idade na qual um coelho pode come√ßar a procriar
    MAX_AGE = 40      # A idade at√© a qual um coelho pode viver
    BREEDING_PROBABILITY = 0.12  # A probabilidade de um coelho procriar
    MAX_LITTER_SIZE = 4          # O n√∫mero m√°ximo de nascimentos

    def __init__(self, random_age, location):
        """
        Cria um novo coelho. Um coelho pode ser criado com idade
        zero (rec√©m-nascido) ou com idade aleat√≥ria.
        """
        self.age = 0
        self.alive = True
        self.location = location

        rand = Randomizer.get_random()
        if random_age:
            self.age = rand.randint(0, self.MAX_AGE - 1)

    def run(self, current_field, next_field_state):
        """
        Isto √© o que o coelho faz na maior parte do tempo - ele corre
        por a√≠. √Äs vezes ele procria ou morre de idade.
        """
        self._increment_age()
        if self.alive:
            free_locations = next_field_state.get_free_adjacent_locations(self.location)
            if free_locations:
                self._give_birth(next_field_state, free_locations)

            # Tenta se mover para uma localiza√ß√£o livre
            if free_locations:
                new_location = free_locations[0]
                self.location = new_location
                next_field_state.place_rabbit(self, new_location)
            else:
                # Superpopula√ß√£o
                self.set_dead()

    def is_alive(self):
        """Verifica se o coelho est√° vivo ou n√£o."""
        return self.alive

    def set_dead(self):
        """
        Indica que o coelho n√£o est√° mais vivo.
        Ele √© removido do campo.
        """
        self.alive = False
        self.location = None

    def get_location(self):
        """Retorna a localiza√ß√£o do coelho."""
        return self.location

    def __str__(self):
        return f"Rabbit(age={self.age}, alive={self.alive}, location={self.location})"

    def _increment_age(self):
        """
        Aumenta a idade.
        Isto pode resultar na morte do coelho.
        """
        self.age += 1
        if self.age > self.MAX_AGE:
            self.set_dead()

    def _give_birth(self, next_field_state, free_locations):
        """
        Verifica se este coelho deve dar √† luz neste passo.
        Novos nascimentos ser√£o feitos em localiza√ß√µes adjacentes livres.
        """
        births = self._breed()
        if births > 0:
            for b in range(births):
                if not free_locations:
                    break
                loc = free_locations.pop(0)
                young = Rabbit(False, loc)
                next_field_state.place_rabbit(young, loc)

    def _breed(self):
        """
        Gera um n√∫mero representando o n√∫mero de nascimentos,
        se pode procriar.
        """
        rand = Randomizer.get_random()
        if self._can_breed() and rand.random() <= self.BREEDING_PROBABILITY:
            return rand.randint(1, self.MAX_LITTER_SIZE)
        else:
            return 0

    def _can_breed(self):
        """Um coelho pode procriar se atingiu a idade de procria√ß√£o."""
        return self.age >= self.BREEDING_AGE


# fox.py
class Fox:
    """
    Um modelo simples de uma raposa.
    Raposas envelhecem, se movem, comem coelhos e morrem.
    """

    # Caracter√≠sticas compartilhadas por todas as raposas (vari√°veis de classe)
    BREEDING_AGE = 15        # A idade na qual uma raposa pode come√ßar a procriar
    MAX_AGE = 150           # A idade at√© a qual uma raposa pode viver
    BREEDING_PROBABILITY = 0.08  # A probabilidade de uma raposa procriar
    MAX_LITTER_SIZE = 2     # O n√∫mero m√°ximo de nascimentos
    RABBIT_FOOD_VALUE = 9   # O valor alimentar de um √∫nico coelho

    def __init__(self, random_age, location):
        """
        Cria uma raposa. Uma raposa pode ser criada como rec√©m-nascida
        (idade zero e sem fome) ou com idade e n√≠vel de fome aleat√≥rios.
        """
        self.alive = True
        self.location = location
        self.age = 0

        rand = Randomizer.get_random()
        if random_age:
            self.age = rand.randint(0, self.MAX_AGE - 1)

        self.food_level = rand.randint(0, self.RABBIT_FOOD_VALUE - 1)

    def hunt(self, current_field, next_field_state):
        """
        Isto √© o que a raposa faz na maior parte do tempo: ela ca√ßa
        coelhos. No processo, ela pode procriar, morrer de fome ou
        morrer de idade.
        """
        self._increment_age()
        self._increment_hunger()

        if self.alive:
            free_locations = next_field_state.get_free_adjacent_locations(self.location)
            if free_locations:
                self._give_birth(next_field_state, free_locations)

            # Move em dire√ß√£o a uma fonte de comida se encontrada
            new_location = self._find_food(current_field)
            if new_location is None and free_locations:
                # Nenhuma comida encontrada - tenta se mover para uma localiza√ß√£o livre
                new_location = free_locations.pop(0)

            # Verifica se foi poss√≠vel se mover
            if new_location is not None:
                self.location = new_location
                next_field_state.place_fox(self, new_location)
            else:
                # Superpopula√ß√£o
                self.set_dead()

    def is_alive(self):
        """Verifica se a raposa est√° viva ou n√£o."""
        return self.alive

    def get_location(self):
        """Retorna a localiza√ß√£o da raposa."""
        return self.location

    def set_dead(self):
        """
        Indica que a raposa n√£o est√° mais viva.
        Ela √© removida do campo.
        """
        self.alive = False
        self.location = None

    def __str__(self):
        return f"Fox(age={self.age}, alive={self.alive}, location={self.location}, food_level={self.food_level})"

    def _increment_age(self):
        """Aumenta a idade. Isto pode resultar na morte da raposa."""
        self.age += 1
        if self.age > self.MAX_AGE:
            self.set_dead()

    def _increment_hunger(self):
        """Deixa esta raposa com mais fome. Isto pode resultar na morte da raposa."""
        self.food_level -= 1
        if self.food_level <= 0:
            self.set_dead()

    def _find_food(self, field):
        """
        Procura por coelhos adjacentes √† localiza√ß√£o atual.
        Apenas o primeiro coelho vivo √© comido.
        """
        adjacent = field.get_adjacent_locations(self.location)

        for loc in adjacent:
            animal = field.get_object_at(loc)
            if isinstance(animal, Rabbit) and animal.is_alive():
                animal.set_dead()
                self.food_level = self.RABBIT_FOOD_VALUE
                return loc

        return None

    def _give_birth(self, next_field_state, free_locations):
        """
        Verifica se esta raposa deve dar √† luz neste passo.
        Novos nascimentos ser√£o feitos em localiza√ß√µes adjacentes livres.
        """
        births = self._breed()
        if births > 0:
            for b in range(births):
                if not free_locations:
                    break
                loc = free_locations.pop(0)
                young = Fox(False, loc)
                next_field_state.place_fox(young, loc)

    def _breed(self):
        """
        Gera um n√∫mero representando o n√∫mero de nascimentos,
        se pode procriar.
        """
        rand = Randomizer.get_random()
        if self._can_breed() and rand.random() <= self.BREEDING_PROBABILITY:
            return rand.randint(1, self.MAX_LITTER_SIZE)
        else:
            return 0

    def _can_breed(self):
        """Uma raposa pode procriar se atingiu a idade de procria√ß√£o."""
        return self.age >= self.BREEDING_AGE


# field.py
class Field:
    """
    Representa uma grade retangular de posi√ß√µes de campo.
    Cada posi√ß√£o √© capaz de armazenar um √∫nico animal/objeto.
    """

    def __init__(self, depth, width):
        """
        Representa um campo com as dimens√µes dadas.
        """
        self.depth = depth
        self.width = width
        self.field = {}  # Animais mapeados por localiza√ß√£o
        self.rabbits = []  # Os dois tipos de animal
        self.foxes = []
        self._rand = Randomizer.get_random()

    def place_rabbit(self, rabbit, location):
        """
        Coloca um coelho na localiza√ß√£o dada.
        Se j√° houver um animal na localiza√ß√£o, ele ser√° perdido.
        """
        assert location is not None
        other = self.field.get(location)

        if isinstance(other, Rabbit):
            self.rabbits.remove(other)
        elif isinstance(other, Fox):
            self.foxes.remove(other)

        self.field[location] = rabbit
        self.rabbits.append(rabbit)

    def place_fox(self, fox, location):
        """
        Coloca uma raposa na localiza√ß√£o dada.
        Se j√° houver um animal na localiza√ß√£o, ele ser√° perdido.
        """
        assert location is not None
        other = self.field.get(location)

        if isinstance(other, Rabbit):
            self.rabbits.remove(other)
        elif isinstance(other, Fox):
            self.foxes.remove(other)

        self.field[location] = fox
        self.foxes.append(fox)


    def get_object_at(self, location):
        """
        Retorna o animal na localiza√ß√£o dada, se houver.
        """
        return self.field.get(location)

    def get_free_adjacent_locations(self, location):
        """
        Obt√©m uma lista embaralhada das localiza√ß√µes adjacentes livres.
        """
        free = []
        adjacent = self.get_adjacent_locations(location)

        for next_location in adjacent:
            animal = self.field.get(next_location)
            if animal is None:
                free.append(next_location)
            else:
                if isinstance(animal, Rabbit) and not animal.is_alive():
                    free.append(next_location)
                elif isinstance(animal, Fox) and not animal.is_alive():
                    free.append(next_location)


        return free


    def get_adjacent_locations(self, location):
        """
        Retorna uma lista embaralhada de localiza√ß√µes adjacentes √† dada.
        A lista n√£o incluir√° a pr√≥pria localiza√ß√£o.
        Todas as localiza√ß√µes estar√£o dentro da grade.
        """
        locations = []
        if location is not None:
            row = location.row
            col = location.col

            for roffset in range(-1, 2):
                next_row = row + roffset
                if 0 <= next_row < self.depth:
                    for coffset in range(-1, 2):
                        next_col = col + coffset
                        # Exclui localiza√ß√µes inv√°lidas e a localiza√ß√£o original
                        if (0 <= next_col < self.width and
                            (roffset != 0 or coffset != 0)):
                            locations.append(Location(next_row, next_col))

            # Embaralha a lista. V√°rios outros m√©todos dependem da lista
            # estar em ordem aleat√≥ria
            self._rand.shuffle(locations) # Corrected line

        return locations

    def field_stats(self):
        """Imprime o n√∫mero de raposas e coelhos no campo."""
        num_foxes = 0
        num_rabbits = 0

        for animal in self.field.values():
            if isinstance(animal, Fox) and animal.is_alive():
                num_foxes += 1
            elif isinstance(animal, Rabbit) and animal.is_alive():
                num_rabbits += 1

        print(f"Rabbits: {num_rabbits} Foxes: {num_foxes}")


    def clear(self):
        """Esvazia o campo."""
        self.field.clear()
        self.rabbits.clear()
        self.foxes.clear()

    def is_viable(self):
        """
        Retorna se h√° pelo menos um coelho e uma raposa no campo.
        """
        has_rabbits = any(rabbit.is_alive() for rabbit in self.rabbits)
        has_foxes = any(fox.is_alive() for fox in self.foxes)
        return has_rabbits and has_foxes

    def get_depth(self):
        """Retorna a profundidade do campo."""
        return self.depth

    def get_width(self):
        """Retorna a largura do campo."""
        return self.width

    def get_rabbits(self):
        """Retorna uma lista dos coelhos que est√£o vivos."""
        return [rabbit for rabbit in self.rabbits if rabbit.is_alive()]

    def get_foxes(self):
        """Retorna uma lista das raposas que est√£o vivas."""
        return [fox for fox in self.foxes if fox.is_alive()]


# field_stats.py
class FieldStats:
    """
    Esta classe coleta e fornece alguns dados estat√≠sticos sobre o estado
    de um campo. √â flex√≠vel: criar√° e manter√° um contador para qualquer
    classe de objeto encontrada no campo.
    """

    def __init__(self):
        """Constr√≥i um objeto FieldStats."""
        # Contadores para cada tipo de entidade (raposa, coelho, etc.) na simula√ß√£o
        self.counters = {}
        # Se os contadores est√£o atualmente atualizados
        self.counts_valid = True


    def get_population_details(self, field):
        """
        Obt√©m detalhes do que est√° no campo.
        :return: Uma string descrevendo o que est√° no campo
        """
        if not self.counts_valid:
            self._generate_counts(field)

        details = []
        for class_type, counter in self.counters.items():
            details.append(f"{counter.get_name()}: {counter.get_count()}")

        return " ".join(details)


    def reset(self):
        """
        Invalida o conjunto atual de estat√≠sticas; redefine todos
        os contadores para zero.
        """
        self.counts_valid = False
        for counter in self.counters.values():
            counter.reset()


    def increment_count(self, animal_class):
        """
        Incrementa o contador para uma classe de animal.
        :param animal_class: A classe do animal a incrementar
        """
        if animal_class not in self.counters:
            # Ainda n√£o temos um contador para esta esp√©cie.
            # Cria um.
            self.counters[animal_class] = Counter(animal_class.__name__)

        self.counters[animal_class].increment()


    def count_finished(self):
        """Indica que uma contagem de animais foi completada."""
        self.counts_valid = True


    def is_viable(self, field):
        """
        Determina se a simula√ß√£o ainda √© vi√°vel.
        Ou seja, se deve continuar a executar.
        :return: true se h√° mais de uma esp√©cie viva
        """
        return self.stats.is_viable(field)


    def _generate_counts(self, field):
        """
        Generates counts of the number of foxes and rabbits.
        These are not kept up to date as foxes and rabbits
        are placed on the field, but only when a request
        is made for the information.
        """
        self.reset()
        for row in range(field.get_depth()):
            for col in range(field.get_width()):
                animal = field.get_object_at(Location(row, col))
                if animal is not None:
                    self.increment_count(type(animal))
        self.counts_valid = True

#simulator_view.py
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np
import os
#from field_stats import FieldStats
#from location import Location
from IPython.display import display, HTML, Image
import time

class SimulatorView:
    """
    Uma visualiza√ß√£o gr√°fica da grade de simula√ß√£o para Google Colab.
    N√£o interfere com o input do usu√°rio.
    """

    def __init__(self, height, width):
        """
        Cria uma visualiza√ß√£o com a altura e largura dadas.
        """
        self.height = height
        self.width = width
        self.stats = FieldStats()

        # Define cores para cada tipo de animal
        self.colors = {
            'Rabbit': 1,  # Verde
            'Fox': 2,     # Vermelho
            'Empty': 0    # Branco/Transparente
        }

        # Configura√ß√£o do matplotlib
        self.colormap = ListedColormap(['white', 'lightgreen', 'red'])

        # Hist√≥rico de popula√ß√µes para gr√°ficos
        self.population_history = {
            'steps': [],
            'rabbits': [],
            'foxes': [],
            'total': []
        }

        # Controla frequ√™ncia de atualiza√ß√£o
        self.update_frequency = 25  # Atualiza gr√°fico a cada 25 steps
        self.print_frequency = 50   # Imprime info a cada 50 steps

        # Configura√ß√£o do matplotlib para n√£o interferir com input
        plt.ioff()  # Desativa modo interativo

        # Cria diret√≥rio para salvar imagens
        self.image_dir = "simulation_images"
        self.setup_image_directory()

        # Cria √°rea de visualiza√ß√£o dedicada
        self.setup_display_area()

    def setup_image_directory(self):
        """Cria diret√≥rio para salvar as imagens da simula√ß√£o."""
        try:
            if not os.path.exists(self.image_dir):
                os.makedirs(self.image_dir)
                print(f"üìÅ Diret√≥rio criado: {self.image_dir}")
            else:
                print(f"üìÅ Usando diret√≥rio existente: {self.image_dir}")
        except Exception as e:
            print(f"‚ö†Ô∏è  Aviso: N√£o foi poss√≠vel criar diret√≥rio {self.image_dir}: {e}")
            # Fallback para diret√≥rio atual
            self.image_dir = "."

    def setup_display_area(self):
        """Configura √°rea dedicada para visualiza√ß√£o."""
        try:
            display(HTML("""
            <div id="simulation-display" style="margin: 20px 0; padding: 10px; border: 1px solid #ddd; border-radius: 5px;">
                <h3>üê∞ü¶ä Simula√ß√£o Predador-Presa</h3>
                <p>A visualiza√ß√£o aparecer√° aqui durante a simula√ß√£o...</p>
            </div>
            """))
        except Exception as e:
            print(f"üì± Modo terminal detectado - visualiza√ß√£o adaptada")

    def set_color(self, animal_class, symbol):
        """
        Define um s√≠mbolo/cor a ser usado para uma dada classe de animal.
        """
        if hasattr(animal_class, '__name__'):
            class_name = animal_class.__name__
        else:
            class_name = str(animal_class)

        # Mapeia s√≠mbolos para c√≥digos de cor
        symbol_to_color = {
            'R': 1,  # Rabbit -> Verde
            'F': 2,  # Fox -> Vermelho
        }

        if symbol in symbol_to_color:
            self.colors[class_name] = symbol_to_color[symbol]

    def show_status(self, step, field):
        """
        Mostra o status atual do campo.
        """
        # Sempre atualiza as estat√≠sticas
        self.stats.reset()

        # Conta popula√ß√µes
        rabbit_count = 0
        fox_count = 0

        for row in range(field.get_depth()):
            for col in range(field.get_width()):
                animal = field.get_object_at(Location(row, col))
                if animal is not None:
                    self.stats.increment_count(type(animal))
                    animal_type = type(animal).__name__

                    if animal_type == 'Rabbit':
                        rabbit_count += 1
                    elif animal_type == 'Fox':
                        fox_count += 1

        self.stats.count_finished()

        # Sempre adiciona ao hist√≥rico
        self.population_history['steps'].append(step)
        self.population_history['rabbits'].append(rabbit_count)
        self.population_history['foxes'].append(fox_count)
        self.population_history['total'].append(rabbit_count + fox_count)

        # Imprime informa√ß√µes de forma compacta
        if step % self.print_frequency == 0:
            print(f"Step {step:4d} | üê∞ {rabbit_count:4d} | ü¶ä {fox_count:3d} | Total: {rabbit_count + fox_count:4d}")

        # Atualiza gr√°ficos periodicamente
        if step % self.update_frequency == 0:
            self.create_and_save_plots(field, step, rabbit_count, fox_count)

    def create_and_save_plots(self, field, step, rabbit_count, fox_count):
        """Cria e salva gr√°ficos da simula√ß√£o."""

        try:
            # Cria figura
            fig, axes = plt.subplots(1, 2, figsize=(15, 6))
            fig.patch.set_facecolor('white')
            fig.suptitle(f"Simula√ß√£o Predador-Presa - Step {step}", fontsize=16, fontweight='bold')

            # Plot 1: Grade do campo (amostra)
            grid_matrix = np.zeros((field.get_depth(), field.get_width()))

            for row in range(field.get_depth()):
                for col in range(field.get_width()):
                    animal = field.get_object_at(Location(row, col))
                    if animal is not None:
                        animal_type = type(animal).__name__
                        if animal_type == 'Rabbit':
                            grid_matrix[row][col] = 1
                        elif animal_type == 'Fox':
                            grid_matrix[row][col] = 2

            # Mostra amostra do campo (limitado para visualiza√ß√£o)
            max_rows = min(40, field.get_depth())
            max_cols = min(60, field.get_width())
            display_grid = grid_matrix[:max_rows, :max_cols]

            im = axes[0].imshow(display_grid, cmap=self.colormap, vmin=0, vmax=2, aspect='equal')
            axes[0].set_title(f"Campo - Step {step}\n Coelhos: {rabbit_count} |  Raposas: {fox_count}")
            axes[0].set_xlabel(f"Largura (mostrando {max_cols}/{field.get_width()})")
            axes[0].set_ylabel(f"Altura (mostrando {max_rows}/{field.get_depth()})")

            # Adiciona legenda de cores
            legend_elements = [
                plt.Rectangle((0,0),1,1, facecolor='lightgreen', label='Coelhos'),
                plt.Rectangle((0,0),1,1, facecolor='red', label='Raposas'),
                plt.Rectangle((0,0),1,1, facecolor='white', edgecolor='black', label='Vazio')
            ]
            axes[0].legend(handles=legend_elements, loc='upper right', bbox_to_anchor=(1.0, 1.0))

            # Adiciona info se o campo foi truncado
            if field.get_depth() > max_rows or field.get_width() > max_cols:
                axes[0].text(0.02, 0.02, f"Amostra: {max_rows}√ó{max_cols}\nCampo real: {field.get_depth()}√ó{field.get_width()}",
                            transform=axes[0].transAxes,
                            verticalalignment='bottom',
                            fontsize=8,
                            bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.7))

            # Plot 2: Evolu√ß√£o populacional
            if len(self.population_history['steps']) > 1:
                axes[1].plot(self.population_history['steps'],
                           self.population_history['rabbits'],
                           'g-', label='Coelhos', linewidth=2.5, marker='o', markersize=3)
                axes[1].plot(self.population_history['steps'],
                           self.population_history['foxes'],
                           'r-', label='Raposas', linewidth=2.5, marker='s', markersize=3)

                # Destaca ponto atual com marcadores maiores
                axes[1].plot(step, rabbit_count, 'go', markersize=8, markeredgecolor='darkgreen', markeredgewidth=2)
                axes[1].plot(step, fox_count, 'ro', markersize=8, markeredgecolor='darkred', markeredgewidth=2)

                # Adiciona texto com valores atuais
                axes[1].text(0.02, 0.98, f"Atual:\n Coelhos {rabbit_count}\nRaposas {fox_count}",
                           transform=axes[1].transAxes,
                           verticalalignment='top',
                           fontsize=10,
                           bbox=dict(boxstyle="round,pad=0.3", facecolor="lightblue", alpha=0.8))

            axes[1].set_title("Evolu√ß√£o Populacional")
            axes[1].set_xlabel("Steps")
            axes[1].set_ylabel("Popula√ß√£o")
            axes[1].legend(loc='upper right')
            axes[1].grid(True, alpha=0.3)

            # Define limites do eixo y para melhor visualiza√ß√£o
            if len(self.population_history['steps']) > 1:
                max_pop = max(max(self.population_history['rabbits']), max(self.population_history['foxes']))
                axes[1].set_ylim(0, max_pop * 1.1)

            plt.tight_layout()

            # Salva a imagem
            filename = f"{self.image_dir}/simulation_step_{step:06d}.png"
            plt.savefig(filename, dpi=150, bbox_inches='tight', facecolor='white')

            print(f"üíæ Gr√°fico salvo: {filename}")

            # Exibe a imagem se estivermos no Colab/Jupyter
            try:
                display(Image(filename))
            except Exception:
                # Se n√£o conseguir exibir, pelo menos confirma que salvou
                print(f"üìä Gr√°fico do Step {step} salvo com sucesso!")

            plt.close(fig)  # Fecha para liberar mem√≥ria

        except Exception as e:
            print(f"‚ùå Erro ao criar/salvar gr√°fico do Step {step}: {e}")
            # Tenta criar pelo menos um gr√°fico simples
            self.create_simple_plot_fallback(step, rabbit_count, fox_count)

    def create_simple_plot_fallback(self, step, rabbit_count, fox_count):
        """Cria um gr√°fico simples como fallback em caso de erro."""
        try:
            fig, ax = plt.subplots(1, 1, figsize=(10, 6))

            if len(self.population_history['steps']) > 1:
                ax.plot(self.population_history['steps'],
                       self.population_history['rabbits'],
                       'g-', label='Coelhos', linewidth=2)
                ax.plot(self.population_history['steps'],
                       self.population_history['foxes'],
                       'r-', label='Raposas', linewidth=2)

            ax.set_title(f"Popula√ß√£o - Step {step}")
            ax.set_xlabel("Steps")
            ax.set_ylabel("Popula√ß√£o")
            ax.legend()
            ax.grid(True, alpha=0.3)

            filename = f"{self.image_dir}/simple_step_{step:06d}.png"
            plt.savefig(filename, dpi=100, bbox_inches='tight')
            plt.close(fig)

            print(f"üíæ Gr√°fico simples salvo: {filename}")

        except Exception as e:
            print(f"‚ùå Erro mesmo no gr√°fico simples: {e}")

    def is_viable(self, field):
        """
        Determina se a simula√ß√£o deve continuar a executar.
        """
        return field.is_viable()

    def set_update_frequency(self, steps):
        """Define frequ√™ncia de atualiza√ß√£o dos gr√°ficos."""
        self.update_frequency = max(1, steps)
        print(f"üìä Gr√°ficos ser√£o salvos a cada {self.update_frequency} steps")

    def set_print_frequency(self, steps):
        """Define frequ√™ncia de informa√ß√µes no console."""
        self.print_frequency = max(1, steps)
        print(f"üìù Console atualizar√° a cada {self.print_frequency} steps")

    def show_final_results(self):
        """Mostra resultados finais da simula√ß√£o."""

        if len(self.population_history['steps']) == 0:
            print("‚ùå Nenhum dado de simula√ß√£o dispon√≠vel.")
            return None

        final_step = self.population_history['steps'][-1]
        final_rabbits = self.population_history['rabbits'][-1]
        final_foxes = self.population_history['foxes'][-1]
        max_rabbits = max(self.population_history['rabbits'])
        max_foxes = max(self.population_history['foxes'])

        # Informa√ß√µes textuais
        print("\n" + "="*60)
        print("üèÅ SIMULA√á√ÉO FINALIZADA")
        print("="*60)
        print(f"üìä Steps executados: {final_step}")
        print(f"üê∞ Coelhos: {final_rabbits} (m√°ximo: {max_rabbits})")
        print(f"ü¶ä Raposas: {final_foxes} (m√°ximo: {max_foxes})")
        print(f"üìà Popula√ß√£o total final: {final_rabbits + final_foxes}")
        print(f"üìÅ Gr√°ficos salvos em: {os.path.abspath(self.image_dir)}")
        print("="*60)

        # Cria gr√°fico final resumo
        self.create_final_summary_plot()

        return {
            'final_step': final_step,
            'final_rabbits': final_rabbits,
            'final_foxes': final_foxes,
            'max_rabbits': max_rabbits,
            'max_foxes': max_foxes,
            'history': self.population_history.copy(),
            'image_directory': os.path.abspath(self.image_dir)
        }

    def create_final_summary_plot(self):
        """Cria gr√°fico final resumo da simula√ß√£o."""
        try:
            fig, ax = plt.subplots(1, 1, figsize=(14, 8))
            fig.patch.set_facecolor('white')

            if len(self.population_history['steps']) > 1:
                ax.plot(self.population_history['steps'],
                       self.population_history['rabbits'],
                       'g-', label='üê∞ Coelhos', linewidth=3, alpha=0.8)
                ax.plot(self.population_history['steps'],
                       self.population_history['foxes'],
                       'r-', label='ü¶ä Raposas', linewidth=3, alpha=0.8)

                # Marca pontos de m√°ximo
                max_rabbits = max(self.population_history['rabbits'])
                max_foxes = max(self.population_history['foxes'])
                max_r_step = self.population_history['steps'][self.population_history['rabbits'].index(max_rabbits)]
                max_f_step = self.population_history['steps'][self.population_history['foxes'].index(max_foxes)]

                ax.plot(max_r_step, max_rabbits, 'go', markersize=12,
                       label=f'Max Coelhos: {max_rabbits} (Step {max_r_step})')
                ax.plot(max_f_step, max_foxes, 'ro', markersize=12,
                       label=f'Max Raposas: {max_foxes} (Step {max_f_step})')

            ax.set_title("üê∞ü¶ä Simula√ß√£o Predador-Presa - Resultado Final",
                        fontsize=16, fontweight='bold')
            ax.set_xlabel("Steps", fontsize=12)
            ax.set_ylabel("Popula√ß√£o", fontsize=12)
            ax.legend(fontsize=11)
            ax.grid(True, alpha=0.3)

            plt.tight_layout()

            # Salva gr√°fico final
            final_filename = f"{self.image_dir}/FINAL_simulation_summary.png"
            plt.savefig(final_filename, dpi=300, bbox_inches='tight', facecolor='white')

            print(f"üìä Resumo final salvo: {final_filename}")

            # Tenta exibir
            try:
                display(Image(final_filename))
                plt.show()
            except:
                print("üì± Gr√°fico salvo - visualize o arquivo para ver o resultado final")

            plt.close(fig)

        except Exception as e:
            print(f"‚ùå Erro ao criar gr√°fico final: {e}")

    def list_saved_images(self):
        """Lista todas as imagens salvas da simula√ß√£o."""
        try:
            files = [f for f in os.listdir(self.image_dir) if f.endswith('.png')]
            files.sort()

            print(f"\nüìÅ Imagens salvas em {os.path.abspath(self.image_dir)}:")
            for i, file in enumerate(files, 1):
                print(f"  {i:3d}. {file}")

            print(f"\nüìä Total: {len(files)} imagens")

        except Exception as e:
            print(f"‚ùå Erro ao listar imagens: {e}")

    def clean_saved_images(self):
        """Remove todas as imagens salvas da simula√ß√£o."""
        try:
            files = [f for f in os.listdir(self.image_dir) if f.endswith('.png')]

            for file in files:
                os.remove(os.path.join(self.image_dir, file))

            print(f"üßπ Removidas {len(files)} imagens do diret√≥rio {self.image_dir}")

        except Exception as e:
            print(f"‚ùå Erro ao limpar imagens: {e}")

# Fun√ß√£o auxiliar otimizada
def create_simulator_view(height=80, width=120, update_freq=25):
    """
    Cria visualizador otimizado que salva gr√°ficos automaticamente.

    Args:
        height: altura do campo (default=80)
        width: largura do campo (default=120)
        update_freq: steps entre salvamentos de gr√°ficos (default=25)

    Returns:
        SimulatorView: objeto visualizador configurado
    """
    view = SimulatorView(height, width)
    view.set_update_frequency(update_freq)

    print("üöÄ Simulador criado!")
    print(f"üìè Campo: {height}√ó{width}")
    print(f"üíæ Gr√°ficos ser√£o salvos a cada {update_freq} steps")
    print(f"üìÅ Diret√≥rio de imagens: {os.path.abspath(view.image_dir)}")
    print("‚ñ∂Ô∏è  Execute simulator.simulate(steps) para come√ßar!")

    return view
'''
# simulator_view.py (vers√£o simplificada para terminal)
class SimulatorView:
    """
    Uma visualiza√ß√£o da grade de simula√ß√£o.
    Esta vers√£o simplificada exibe informa√ß√µes no terminal.
    """

    def __init__(self, height, width):
        """
        Cria uma visualiza√ß√£o com a altura e largura dadas.
        """
        self.height = height
        self.width = width
        self.stats = FieldStats()

        # Define cores para cada tipo de animal
        self.colors = {
            'Rabbit': 'R',
            'Fox': 'F'
        }

    def set_color(self, animal_class, symbol):
        """
        Define um s√≠mbolo a ser usado para uma dada classe de animal.
        """
        self.colors[animal_class.__name__] = symbol

    def show_status(self, step, field):
        """
        Mostra o status atual do campo.
        """
        print(f"\n=== Step: {step} ===")

        self.stats.reset()

        # Cria uma representa√ß√£o visual do campo
        grid = [['.' for _ in range(field.get_width())]
                for _ in range(field.get_depth())]

        for row in range(field.get_depth()):
            for col in range(field.get_width()):
                animal = field.get_object_at(Location(row, col))
                if animal is not None:
                    self.stats.increment_count(type(animal))
                    symbol = self.colors.get(type(animal).__name__, '?')
                    grid[row][col] = symbol

        self.stats.count_finished()

        # Exibe apenas uma amostra do grid (primeiras 20 linhas e colunas)
        max_display_rows = min(20, field.get_depth())
        max_display_cols = min(40, field.get_width())

        for row in range(max_display_rows):
            print(''.join(grid[row][:max_display_cols]))

        if field.get_depth() > max_display_rows or field.get_width() > max_display_cols:
            print("... (campo truncado para visualiza√ß√£o)")

        print(f"Population: {self.stats.get_population_details(field)}")

    def is_viable(self, field):
        """
        Determina se a simula√ß√£o deve continuar a executar.
        """
        return self.stats.is_viable(field)

'''
# simulator.py
class Simulator:
    """
    Um simulador simples predador-presa, baseado em um campo retangular
    contendo coelhos e raposas.
    """

    # Constantes representando informa√ß√µes de configura√ß√£o para a simula√ß√£o
    DEFAULT_WIDTH = 120   # A largura padr√£o da grade
    DEFAULT_DEPTH = 80    # A profundidade padr√£o da grade
    FOX_CREATION_PROBABILITY = 0.02      # A probabilidade de uma raposa ser criada
    RABBIT_CREATION_PROBABILITY = 0.08   # A probabilidade de um coelho ser criado

    def __init__(self, depth=None, width=None):
        """
        Cria um campo de simula√ß√£o com o tamanho dado.
        """
        if depth is None:
            depth = self.DEFAULT_DEPTH
        if width is None:
            width = self.DEFAULT_WIDTH

        if width <= 0 or depth <= 0:
            print("As dimens√µes devem ser >= zero.")
            print("Usando valores padr√£o.")
            depth = self.DEFAULT_DEPTH
            width = self.DEFAULT_WIDTH

        self.field = Field(depth, width)
        self.view = SimulatorView(depth, width)
        self.step = 0

        self.reset()

    def run_long_simulation(self):
        """
        Executa a simula√ß√£o a partir de seu estado atual por um
        per√≠odo razoavelmente longo (4000 passos).
        """
        self.simulate(4000)

    def simulate(self, num_steps):
        """
        Executa a simula√ß√£o pelo n√∫mero dado de passos.
        Para antes do n√∫mero dado de passos se deixar de ser vi√°vel.
        """
        self.report_stats()
        for n in range(1, num_steps + 1):
            if not self.field.is_viable():
                break
            self.simulate_one_step()
            self._delay(100)  # ajuste isto para mudar a velocidade de execu√ß√£o

    def simulate_one_step(self):
        """
        Executa a simula√ß√£o a partir de seu estado atual por um √∫nico passo.
        Itera sobre todo o campo atualizando o estado de cada raposa e coelho.
        """
        self.step += 1
        # Usa um Field separado para armazenar o estado inicial do pr√≥ximo passo
        next_field_state = Field(self.field.get_depth(), self.field.get_width())

        rabbits = self.field.get_rabbits()
        foxes = self.field.get_foxes()

        # Deixa todos os coelhos correrem
        for rabbit in rabbits:
            rabbit.run(self.field, next_field_state)

        # Deixa todas as raposas ca√ßarem
        for fox in foxes:
            fox.hunt(self.field, next_field_state)

        # Substitui o estado antigo pelo novo
        self.field = next_field_state

        self.report_stats()
        self.view.show_status(self.step, self.field)

    def reset(self):
        """Redefine a simula√ß√£o para uma posi√ß√£o inicial."""
        self.step = 0
        self._populate()
        self.view.show_status(self.step, self.field)

    def report_stats(self):
        """Relata o n√∫mero de cada tipo de animal no campo."""
        print(f"Step: {self.step} ", end="")
        self.field.field_stats()

    def _populate(self):
        """Popula aleatoriamente o campo com raposas e coelhos."""
        rand = Randomizer.get_random()
        self.field.clear()

        for row in range(self.field.get_depth()):
            for col in range(self.field.get_width()):
                if rand.random() <= self.FOX_CREATION_PROBABILITY:
                    location = Location(row, col)
                    fox = Fox(True, location)
                    self.field.place_fox(fox, location)
                elif rand.random() <= self.RABBIT_CREATION_PROBABILITY:
                    location = Location(row, col)
                    rabbit = Rabbit(True, location)
                    self.field.place_rabbit(rabbit, location)
                # sen√£o deixa a localiza√ß√£o vazia

    def _delay(self, milliseconds):
        """
        Pausa por um tempo dado.
        :param milliseconds: O tempo para pausar, em milissegundos
        """
        time.sleep(milliseconds / 1000.0)


# main.py - exemplo de uso
if __name__ == "__main__":
    print("=== Simula√ß√£o Foxes and Rabbits ===")
    print("Comandos dispon√≠veis:")
    print("1 - Simular um passo")
    print("2 - Simular N passos")
    print("3 - Executar simula√ß√£o longa (4000 passos)")
    print("4 - Resetar simula√ß√£o")
    print("0 - Sair")

    # Cria o simulador
    simulator = Simulator(40, 60)  # Campo menor para melhor visualiza√ß√£o

    while True:
        try:
            choice = input("\nEscolha uma op√ß√£o: ").strip()
            if choice == "0":
                print("Encerrando simula√ß√£o...")
                break
            elif choice == "1":
                simulator.simulate_one_step()
            elif choice == "2":
                try:
                    steps = int(input("N√∫mero de passos: "))
                    if steps > 0:
                        simulator.simulate(steps)
                    else:
                        print("N√∫mero de passos deve ser positivo.")
                except ValueError:
                    print("Por favor, digite um n√∫mero v√°lido.")
            elif choice == "3":
                print("Iniciando simula√ß√£o longa...")
                simulator.run_long_simulation()
            elif choice == "4":
                simulator.reset()
                print("Simula√ß√£o resetada.")
            else:
                print("Op√ß√£o inv√°lida. Tente novamente.")

        except KeyboardInterrupt:
            print("\n\nSimula√ß√£o interrompida pelo usu√°rio.")
            break
        except Exception as e:
            print(f"Erro: {e}")

    print("Simula√ß√£o finalizada.")
```

```python
%%writefile location.py

# location.py
class Location:
    """Representa uma localiza√ß√£o em uma grade retangular."""

    def __init__(self, row, col):
        self.row = row
        self.col = col

    def __eq__(self, other):
        if isinstance(other, Location):
            return self.row == other.row and self.col == other.col
        return False

    def __hash__(self):
        return hash((self.row, self.col))

    def __str__(self):
        return f"Location({self.row}, {self.col})"

    def __repr__(self):
        return self.__str__()
```

```python
%%writefile randomizer.py
import random
# randomizer.py
class Randomizer:
    """
    Fornece controle sobre a aleatoriedade da simula√ß√£o. Usando o
    randomizador compartilhado com semente fixa, execu√ß√µes repetidas
    ter√£o exatamente o mesmo comportamento (√∫til para testes).
    """

    # A semente padr√£o para controle da aleatoriedade
    SEED = 1111
    # Um objeto Random compartilhado, se necess√°rio
    _rand = random.Random(SEED)
    # Determina se um gerador aleat√≥rio compartilhado deve ser fornecido
    USE_SHARED = True

    @staticmethod
    def get_random():
        """Fornece um gerador aleat√≥rio."""
        if Randomizer.USE_SHARED:
            return Randomizer._rand
        else:
            return random.Random()

    @staticmethod
    def reset():
        """
        Redefine a aleatoriedade.
        N√£o ter√° efeito se a aleatoriedade n√£o for atrav√©s de um
        gerador Random compartilhado.
        """
        if Randomizer.USE_SHARED:
            Randomizer._rand.seed(Randomizer.SEED)
```

```python
%%writefile rabbit.py
from randomizer import Randomizer

# rabbit.py
class Rabbit:
    """
    Um modelo simples de um coelho.
    Coelhos envelhecem, se movem, procriam e morrem.
    """

    # Caracter√≠sticas compartilhadas por todos os coelhos (vari√°veis de classe)
    BREEDING_AGE = 5  # A idade na qual um coelho pode come√ßar a procriar
    MAX_AGE = 40      # A idade at√© a qual um coelho pode viver
    BREEDING_PROBABILITY = 0.12  # A probabilidade de um coelho procriar
    MAX_LITTER_SIZE = 4          # O n√∫mero m√°ximo de nascimentos

    def __init__(self, random_age, location):
        """
        Cria um novo coelho. Um coelho pode ser criado com idade
        zero (rec√©m-nascido) ou com idade aleat√≥ria.
        """
        self.age = 0
        self.alive = True
        self.location = location

        rand = Randomizer.get_random()
        if random_age:
            self.age = rand.randint(0, self.MAX_AGE - 1)

    def run(self, current_field, next_field_state):
        """
        Isto √© o que o coelho faz na maior parte do tempo - ele corre
        por a√≠. √Äs vezes ele procria ou morre de idade.
        """
        self._increment_age()
        if self.alive:
            free_locations = next_field_state.get_free_adjacent_locations(self.location)
            if free_locations:
                self._give_birth(next_field_state, free_locations)

            # Tenta se mover para uma localiza√ß√£o livre
            if free_locations:
                new_location = free_locations[0]
                self.location = new_location
                next_field_state.place_rabbit(self, new_location)
            else:
                # Superpopula√ß√£o
                self.set_dead()

    def is_alive(self):
        """Verifica se o coelho est√° vivo ou n√£o."""
        return self.alive

    def set_dead(self):
        """
        Indica que o coelho n√£o est√° mais vivo.
        Ele √© removido do campo.
        """
        self.alive = False
        self.location = None

    def get_location(self):
        """Retorna a localiza√ß√£o do coelho."""
        return self.location

    def __str__(self):
        return f"Rabbit(age={self.age}, alive={self.alive}, location={self.location})"

    def _increment_age(self):
        """
        Aumenta a idade.
        Isto pode resultar na morte do coelho.
        """
        self.age += 1
        if self.age > self.MAX_AGE:
            self.set_dead()

    def _give_birth(self, next_field_state, free_locations):
        """
        Verifica se este coelho deve dar √† luz neste passo.
        Novos nascimentos ser√£o feitos em localiza√ß√µes adjacentes livres.
        """
        births = self._breed()
        if births > 0:
            for b in range(births):
                if not free_locations:
                    break
                loc = free_locations.pop(0)
                young = Rabbit(False, loc)
                next_field_state.place_rabbit(young, loc)

    def _breed(self):
        """
        Gera um n√∫mero representando o n√∫mero de nascimentos,
        se pode procriar.
        """
        rand = Randomizer.get_random()
        if self._can_breed() and rand.random() <= self.BREEDING_PROBABILITY:
            return rand.randint(1, self.MAX_LITTER_SIZE)
        else:
            return 0

    def _can_breed(self):
        """Um coelho pode procriar se atingiu a idade de procria√ß√£o."""
        return self.age >= self.BREEDING_AGE
```

```python
%%writefile fox.py

from randomizer import Randomizer
# fox.py
class Fox:
    """
    Um modelo simples de uma raposa.
    Raposas envelhecem, se movem, comem coelhos e morrem.
    """

    # Caracter√≠sticas compartilhadas por todas as raposas (vari√°veis de classe)
    BREEDING_AGE = 15        # A idade na qual uma raposa pode come√ßar a procriar
    MAX_AGE = 150           # A idade at√© a qual uma raposa pode viver
    BREEDING_PROBABILITY = 0.08  # A probabilidade de uma raposa procriar
    MAX_LITTER_SIZE = 2     # O n√∫mero m√°ximo de nascimentos
    RABBIT_FOOD_VALUE = 9   # O valor alimentar de um √∫nico coelho

    def __init__(self, random_age, location):
        """
        Cria uma raposa. Uma raposa pode ser criada como rec√©m-nascida
        (idade zero e sem fome) ou com idade e n√≠vel de fome aleat√≥rios.
        """
        self.alive = True
        self.location = location
        self.age = 0

        rand = Randomizer.get_random()
        if random_age:
            self.age = rand.randint(0, self.MAX_AGE - 1)

        self.food_level = rand.randint(0, self.RABBIT_FOOD_VALUE - 1)

    def hunt(self, current_field, next_field_state):
        """
        Isto √© o que a raposa faz na maior parte do tempo: ela ca√ßa
        coelhos. No processo, ela pode procriar, morrer de fome ou
        morrer de idade.
        """
        self._increment_age()
        self._increment_hunger()

        if self.alive:
            free_locations = next_field_state.get_free_adjacent_locations(self.location)
            if free_locations:
                self._give_birth(next_field_state, free_locations)

            # Move em dire√ß√£o a uma fonte de comida se encontrada
            new_location = self._find_food(current_field)
            if new_location is None and free_locations:
                # Nenhuma comida encontrada - tenta se mover para uma localiza√ß√£o livre
                new_location = free_locations.pop(0)

            # Verifica se foi poss√≠vel se mover
            if new_location is not None:
                self.location = new_location
                next_field_state.place_fox(self, new_location)
            else:
                # Superpopula√ß√£o
                self.set_dead()

    def is_alive(self):
        """Verifica se a raposa est√° viva ou n√£o."""
        return self.alive

    def get_location(self):
        """Retorna a localiza√ß√£o da raposa."""
        return self.location

    def set_dead(self):
        """
        Indica que a raposa n√£o est√° mais viva.
        Ela √© removida do campo.
        """
        self.alive = False
        self.location = None

    def __str__(self):
        return f"Fox(age={self.age}, alive={self.alive}, location={self.location}, food_level={self.food_level})"

    def _increment_age(self):
        """Aumenta a idade. Isto pode resultar na morte da raposa."""
        self.age += 1
        if self.age > self.MAX_AGE:
            self.set_dead()

    def _increment_hunger(self):
        """Deixa esta raposa com mais fome. Isto pode resultar na morte da raposa."""
        self.food_level -= 1
        if self.food_level <= 0:
            self.set_dead()

    def _find_food(self, field):
        """
        Procura por coelhos adjacentes √† localiza√ß√£o atual.
        Apenas o primeiro coelho vivo √© comido.
        """
        adjacent = field.get_adjacent_locations(self.location)

        for loc in adjacent:
            animal = field.get_object_at(loc)
            if isinstance(animal, Rabbit) and animal.is_alive():
                animal.set_dead()
                self.food_level = self.RABBIT_FOOD_VALUE
                return loc

        return None

    def _give_birth(self, next_field_state, free_locations):
        """
        Verifica se esta raposa deve dar √† luz neste passo.
        Novos nascimentos ser√£o feitos em localiza√ß√µes adjacentes livres.
        """
        births = self._breed()
        if births > 0:
            for b in range(births):
                if not free_locations:
                    break
                loc = free_locations.pop(0)
                young = Fox(False, loc)
                next_field_state.place_fox(young, loc)

    def _breed(self):
        """
        Gera um n√∫mero representando o n√∫mero de nascimentos,
        se pode procriar.
        """
        rand = Randomizer.get_random()
        if self._can_breed() and rand.random() <= self.BREEDING_PROBABILITY:
            return rand.randint(1, self.MAX_LITTER_SIZE)
        else:
            return 0

    def _can_breed(self):
        """Uma raposa pode procriar se atingiu a idade de procria√ß√£o."""
        return self.age >= self.BREEDING_AGE
```

```python
%%writefile field.py


# field.py
class Field:
    """
    Representa uma grade retangular de posi√ß√µes de campo.
    Cada posi√ß√£o √© capaz de armazenar um √∫nico animal/objeto.
    """

    def __init__(self, depth, width):
        """
        Representa um campo com as dimens√µes dadas.
        """
        self.depth = depth
        self.width = width
        self.field = {}  # Animais mapeados por localiza√ß√£o
        self.rabbits = []  # Os dois tipos de animal
        self.foxes = []
        self._rand = Randomizer.get_random()

    def place_rabbit(self, rabbit, location):
        """
        Coloca um coelho na localiza√ß√£o dada.
        Se j√° houver um animal na localiza√ß√£o, ele ser√° perdido.
        """
        assert location is not None
        other = self.field.get(location)

        if isinstance(other, Rabbit):
            self.rabbits.remove(other)
        elif isinstance(other, Fox):
            self.foxes.remove(other)

        self.field[location] = rabbit
        self.rabbits.append(rabbit)

    def place_fox(self, fox, location):
        """
        Coloca uma raposa na localiza√ß√£o dada.
        Se j√° houver um animal na localiza√ß√£o, ele ser√° perdido.
        """
        assert location is not None
        other = self.field.get(location)

        if isinstance(other, Rabbit):
            self.rabbits.remove(other)
        elif isinstance(other, Fox):
            self.foxes.remove(other)

        self.field[location] = fox
        self.foxes.append(fox)


    def get_object_at(self, location):
        """
        Retorna o animal na localiza√ß√£o dada, se houver.
        """
        return self.field.get(location)

    def get_free_adjacent_locations(self, location):
        """
        Obt√©m uma lista embaralhada das localiza√ß√µes adjacentes livres.
        """
        free = []
        adjacent = self.get_adjacent_locations(location)

        for next_location in adjacent:
            animal = self.field.get(next_location)
            if animal is None:
                free.append(next_location)
            else:
                if isinstance(animal, Rabbit) and not animal.is_alive():
                    free.append(next_location)
                elif isinstance(animal, Fox) and not animal.is_alive():
                    free.append(next_location)


        return free


    def get_adjacent_locations(self, location):
        """
        Retorna uma lista embaralhada de localiza√ß√µes adjacentes √† dada.
        A lista n√£o incluir√° a pr√≥pria localiza√ß√£o.
        Todas as localiza√ß√µes estar√£o dentro da grade.
        """
        locations = []
        if location is not None:
            row = location.row
            col = location.col

            for roffset in range(-1, 2):
                next_row = row + roffset
                if 0 <= next_row < self.depth:
                    for coffset in range(-1, 2):
                        next_col = col + coffset
                        # Exclui localiza√ß√µes inv√°lidas e a localiza√ß√£o original
                        if (0 <= next_col < self.width and
                            (roffset != 0 or coffset != 0)):
                            locations.append(Location(next_row, next_col))

            # Embaralha a lista. V√°rios outros m√©todos dependem da lista
            # estar em ordem aleat√≥ria
            self._rand.shuffle(locations) # Corrected line

        return locations

    def field_stats(self):
        """Imprime o n√∫mero de raposas e coelhos no campo."""
        num_foxes = 0
        num_rabbits = 0

        for animal in self.field.values():
            if isinstance(animal, Fox) and animal.is_alive():
                num_foxes += 1
            elif isinstance(animal, Rabbit) and animal.is_alive():
                num_rabbits += 1

        print(f"Rabbits: {num_rabbits} Foxes: {num_foxes}")


    def clear(self):
        """Esvazia o campo."""
        self.field.clear()
        self.rabbits.clear()
        self.foxes.clear()

    def is_viable(self):
        """
        Retorna se h√° pelo menos um coelho e uma raposa no campo.
        """
        has_rabbits = any(rabbit.is_alive() for rabbit in self.rabbits)
        has_foxes = any(fox.is_alive() for fox in self.foxes)
        return has_rabbits and has_foxes

    def get_depth(self):
        """Retorna a profundidade do campo."""
        return self.depth

    def get_width(self):
        """Retorna a largura do campo."""
        return self.width

    def get_rabbits(self):
        """Retorna uma lista dos coelhos que est√£o vivos."""
        return [rabbit for rabbit in self.rabbits if rabbit.is_alive()]

    def get_foxes(self):
        """Retorna uma lista das raposas que est√£o vivas."""
        return [fox for fox in self.foxes if fox.is_alive()]
```

```python
%%writefile field_stats.py

# field_stats.py
class FieldStats:
    """
    Esta classe coleta e fornece alguns dados estat√≠sticos sobre o estado
    de um campo. √â flex√≠vel: criar√° e manter√° um contador para qualquer
    classe de objeto encontrada no campo.
    """

    def __init__(self):
        """Constr√≥i um objeto FieldStats."""
        # Contadores para cada tipo de entidade (raposa, coelho, etc.) na simula√ß√£o
        self.counters = {}
        # Se os contadores est√£o atualmente atualizados
        self.counts_valid = True


    def get_population_details(self, field):
        """
        Obt√©m detalhes do que est√° no campo.
        :return: Uma string descrevendo o que est√° no campo
        """
        if not self.counts_valid:
            self._generate_counts(field)

        details = []
        for class_type, counter in self.counters.items():
            details.append(f"{counter.get_name()}: {counter.get_count()}")

        return " ".join(details)


    def reset(self):
        """
        Invalida o conjunto atual de estat√≠sticas; redefine todos
        os contadores para zero.
        """
        self.counts_valid = False
        for counter in self.counters.values():
            counter.reset()


    def increment_count(self, animal_class):
        """
        Incrementa o contador para uma classe de animal.
        :param animal_class: A classe do animal a incrementar
        """
        if animal_class not in self.counters:
            # Ainda n√£o temos um contador para esta esp√©cie.
            # Cria um.
            self.counters[animal_class] = Counter(animal_class.__name__)

        self.counters[animal_class].increment()


    def count_finished(self):
        """Indica que uma contagem de animais foi completada."""
        self.counts_valid = True


    def is_viable(self, field):
        """
        Determina se a simula√ß√£o ainda √© vi√°vel.
        Ou seja, se deve continuar a executar.
        :return: true se h√° mais de uma esp√©cie viva
        """
        return self.stats.is_viable(field)


    def _generate_counts(self, field):
        """
        Generates counts of the number of foxes and rabbits.
        These are not kept up to date as foxes and rabbits
        are placed on the field, but only when a request
        is made for the information.
        """
        self.reset()
        for row in range(field.get_depth()):
            for col in range(field.get_width()):
                animal = field.get_object_at(Location(row, col))
                if animal is not None:
                    self.increment_count(type(animal))
        self.counts_valid = True
```

```python
%%writefile simulator_view.py

#simulator_view.py
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np
import os
from field_stats import FieldStats
from location import Location
from IPython.display import display, HTML, Image
import time

class SimulatorView:
    """
    Uma visualiza√ß√£o gr√°fica da grade de simula√ß√£o para Google Colab.
    N√£o interfere com o input do usu√°rio.
    """

    def __init__(self, height, width):
        """
        Cria uma visualiza√ß√£o com a altura e largura dadas.
        """
        self.height = height
        self.width = width
        self.stats = FieldStats()

        # Define cores para cada tipo de animal
        self.colors = {
            'Rabbit': 1,  # Verde
            'Fox': 2,     # Vermelho
            'Empty': 0    # Branco/Transparente
        }

        # Configura√ß√£o do matplotlib
        self.colormap = ListedColormap(['white', 'lightgreen', 'red'])

        # Hist√≥rico de popula√ß√µes para gr√°ficos
        self.population_history = {
            'steps': [],
            'rabbits': [],
            'foxes': [],
            'total': []
        }

        # Controla frequ√™ncia de atualiza√ß√£o
        self.update_frequency = 25  # Atualiza gr√°fico a cada 25 steps
        self.print_frequency = 50   # Imprime info a cada 50 steps

        # Configura√ß√£o do matplotlib para n√£o interferir com input
        plt.ioff()  # Desativa modo interativo

        # Cria diret√≥rio para salvar imagens
        self.image_dir = "simulation_images"
        self.setup_image_directory()

        # Cria √°rea de visualiza√ß√£o dedicada
        self.setup_display_area()

    def setup_image_directory(self):
        """Cria diret√≥rio para salvar as imagens da simula√ß√£o."""
        try:
            if not os.path.exists(self.image_dir):
                os.makedirs(self.image_dir)
                print(f"üìÅ Diret√≥rio criado: {self.image_dir}")
            else:
                print(f"üìÅ Usando diret√≥rio existente: {self.image_dir}")
        except Exception as e:
            print(f"‚ö†Ô∏è  Aviso: N√£o foi poss√≠vel criar diret√≥rio {self.image_dir}: {e}")
            # Fallback para diret√≥rio atual
            self.image_dir = "."

    def setup_display_area(self):
        """Configura √°rea dedicada para visualiza√ß√£o."""
        try:
            display(HTML("""
            <div id="simulation-display" style="margin: 20px 0; padding: 10px; border: 1px solid #ddd; border-radius: 5px;">
                <h3>üê∞ü¶ä Simula√ß√£o Predador-Presa</h3>
                <p>A visualiza√ß√£o aparecer√° aqui durante a simula√ß√£o...</p>
            </div>
            """))
        except Exception as e:
            print(f"üì± Modo terminal detectado - visualiza√ß√£o adaptada")

    def set_color(self, animal_class, symbol):
        """
        Define um s√≠mbolo/cor a ser usado para uma dada classe de animal.
        """
        if hasattr(animal_class, '__name__'):
            class_name = animal_class.__name__
        else:
            class_name = str(animal_class)

        # Mapeia s√≠mbolos para c√≥digos de cor
        symbol_to_color = {
            'R': 1,  # Rabbit -> Verde
            'F': 2,  # Fox -> Vermelho
        }

        if symbol in symbol_to_color:
            self.colors[class_name] = symbol_to_color[symbol]

    def show_status(self, step, field):
        """
        Mostra o status atual do campo.
        """
        # Sempre atualiza as estat√≠sticas
        self.stats.reset()

        # Conta popula√ß√µes
        rabbit_count = 0
        fox_count = 0

        for row in range(field.get_depth()):
            for col in range(field.get_width()):
                animal = field.get_object_at(Location(row, col))
                if animal is not None:
                    self.stats.increment_count(type(animal))
                    animal_type = type(animal).__name__

                    if animal_type == 'Rabbit':
                        rabbit_count += 1
                    elif animal_type == 'Fox':
                        fox_count += 1

        self.stats.count_finished()

        # Sempre adiciona ao hist√≥rico
        self.population_history['steps'].append(step)
        self.population_history['rabbits'].append(rabbit_count)
        self.population_history['foxes'].append(fox_count)
        self.population_history['total'].append(rabbit_count + fox_count)

        # Imprime informa√ß√µes de forma compacta
        if step % self.print_frequency == 0:
            print(f"Step {step:4d} | üê∞ {rabbit_count:4d} | ü¶ä {fox_count:3d} | Total: {rabbit_count + fox_count:4d}")

        # Atualiza gr√°ficos periodicamente
        if step % self.update_frequency == 0:
            self.create_and_save_plots(field, step, rabbit_count, fox_count)

    def create_and_save_plots(self, field, step, rabbit_count, fox_count):
        """Cria e salva gr√°ficos da simula√ß√£o."""

        try:
            # Cria figura
            fig, axes = plt.subplots(1, 2, figsize=(15, 6))
            fig.patch.set_facecolor('white')
            fig.suptitle(f"Simula√ß√£o Predador-Presa - Step {step}", fontsize=16, fontweight='bold')

            # Plot 1: Grade do campo (amostra)
            grid_matrix = np.zeros((field.get_depth(), field.get_width()))

            for row in range(field.get_depth()):
                for col in range(field.get_width()):
                    animal = field.get_object_at(Location(row, col))
                    if animal is not None:
                        animal_type = type(animal).__name__
                        if animal_type == 'Rabbit':
                            grid_matrix[row][col] = 1
                        elif animal_type == 'Fox':
                            grid_matrix[row][col] = 2

            # Mostra amostra do campo (limitado para visualiza√ß√£o)
            max_rows = min(40, field.get_depth())
            max_cols = min(60, field.get_width())
            display_grid = grid_matrix[:max_rows, :max_cols]

            im = axes[0].imshow(display_grid, cmap=self.colormap, vmin=0, vmax=2, aspect='equal')
            axes[0].set_title(f"Campo - Step {step}\n Coelhos: {rabbit_count} |  Raposas: {fox_count}")
            axes[0].set_xlabel(f"Largura (mostrando {max_cols}/{field.get_width()})")
            axes[0].set_ylabel(f"Altura (mostrando {max_rows}/{field.get_depth()})")

            # Adiciona legenda de cores
            legend_elements = [
                plt.Rectangle((0,0),1,1, facecolor='lightgreen', label='Coelhos'),
                plt.Rectangle((0,0),1,1, facecolor='red', label='Raposas'),
                plt.Rectangle((0,0),1,1, facecolor='white', edgecolor='black', label='Vazio')
            ]
            axes[0].legend(handles=legend_elements, loc='upper right', bbox_to_anchor=(1.0, 1.0))

            # Adiciona info se o campo foi truncado
            if field.get_depth() > max_rows or field.get_width() > max_cols:
                axes[0].text(0.02, 0.02, f"Amostra: {max_rows}√ó{max_cols}\nCampo real: {field.get_depth()}√ó{field.get_width()}",
                            transform=axes[0].transAxes,
                            verticalalignment='bottom',
                            fontsize=8,
                            bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.7))

            # Plot 2: Evolu√ß√£o populacional
            if len(self.population_history['steps']) > 1:
                axes[1].plot(self.population_history['steps'],
                           self.population_history['rabbits'],
                           'g-', label='Coelhos', linewidth=2.5, marker='o', markersize=3)
                axes[1].plot(self.population_history['steps'],
                           self.population_history['foxes'],
                           'r-', label='Raposas', linewidth=2.5, marker='s', markersize=3)

                # Destaca ponto atual com marcadores maiores
                axes[1].plot(step, rabbit_count, 'go', markersize=8, markeredgecolor='darkgreen', markeredgewidth=2)
                axes[1].plot(step, fox_count, 'ro', markersize=8, markeredgecolor='darkred', markeredgewidth=2)

                # Adiciona texto com valores atuais
                axes[1].text(0.02, 0.98, f"Atual:\n Coelhos {rabbit_count}\nRaposas {fox_count}",
                           transform=axes[1].transAxes,
                           verticalalignment='top',
                           fontsize=10,
                           bbox=dict(boxstyle="round,pad=0.3", facecolor="lightblue", alpha=0.8))

            axes[1].set_title("Evolu√ß√£o Populacional")
            axes[1].set_xlabel("Steps")
            axes[1].set_ylabel("Popula√ß√£o")
            axes[1].legend(loc='upper right')
            axes[1].grid(True, alpha=0.3)

            # Define limites do eixo y para melhor visualiza√ß√£o
            if len(self.population_history['steps']) > 1:
                max_pop = max(max(self.population_history['rabbits']), max(self.population_history['foxes']))
                axes[1].set_ylim(0, max_pop * 1.1)

            plt.tight_layout()

            # Salva a imagem
            filename = f"{self.image_dir}/simulation_step_{step:06d}.png"
            plt.savefig(filename, dpi=150, bbox_inches='tight', facecolor='white')

            print(f"üíæ Gr√°fico salvo: {filename}")

            # Exibe a imagem se estivermos no Colab/Jupyter
            try:
                display(Image(filename))
            except Exception:
                # Se n√£o conseguir exibir, pelo menos confirma que salvou
                print(f"üìä Gr√°fico do Step {step} salvo com sucesso!")

            plt.close(fig)  # Fecha para liberar mem√≥ria

        except Exception as e:
            print(f"‚ùå Erro ao criar/salvar gr√°fico do Step {step}: {e}")
            # Tenta criar pelo menos um gr√°fico simples
            self.create_simple_plot_fallback(step, rabbit_count, fox_count)

    def create_simple_plot_fallback(self, step, rabbit_count, fox_count):
        """Cria um gr√°fico simples como fallback em caso de erro."""
        try:
            fig, ax = plt.subplots(1, 1, figsize=(10, 6))

            if len(self.population_history['steps']) > 1:
                ax.plot(self.population_history['steps'],
                       self.population_history['rabbits'],
                       'g-', label='Coelhos', linewidth=2)
                ax.plot(self.population_history['steps'],
                       self.population_history['foxes'],
                       'r-', label='Raposas', linewidth=2)

            ax.set_title(f"Popula√ß√£o - Step {step}")
            ax.set_xlabel("Steps")
            ax.set_ylabel("Popula√ß√£o")
            ax.legend()
            ax.grid(True, alpha=0.3)

            filename = f"{self.image_dir}/simple_step_{step:06d}.png"
            plt.savefig(filename, dpi=100, bbox_inches='tight')
            plt.close(fig)

            print(f"üíæ Gr√°fico simples salvo: {filename}")

        except Exception as e:
            print(f"‚ùå Erro mesmo no gr√°fico simples: {e}")

    def is_viable(self, field):
        """
        Determina se a simula√ß√£o deve continuar a executar.
        """
        return field.is_viable()

    def set_update_frequency(self, steps):
        """Define frequ√™ncia de atualiza√ß√£o dos gr√°ficos."""
        self.update_frequency = max(1, steps)
        print(f"üìä Gr√°ficos ser√£o salvos a cada {self.update_frequency} steps")

    def set_print_frequency(self, steps):
        """Define frequ√™ncia de informa√ß√µes no console."""
        self.print_frequency = max(1, steps)
        print(f"üìù Console atualizar√° a cada {self.print_frequency} steps")

    def show_final_results(self):
        """Mostra resultados finais da simula√ß√£o."""

        if len(self.population_history['steps']) == 0:
            print("‚ùå Nenhum dado de simula√ß√£o dispon√≠vel.")
            return None

        final_step = self.population_history['steps'][-1]
        final_rabbits = self.population_history['rabbits'][-1]
        final_foxes = self.population_history['foxes'][-1]
        max_rabbits = max(self.population_history['rabbits'])
        max_foxes = max(self.population_history['foxes'])

        # Informa√ß√µes textuais
        print("\n" + "="*60)
        print("üèÅ SIMULA√á√ÉO FINALIZADA")
        print("="*60)
        print(f"üìä Steps executados: {final_step}")
        print(f"üê∞ Coelhos: {final_rabbits} (m√°ximo: {max_rabbits})")
        print(f"ü¶ä Raposas: {final_foxes} (m√°ximo: {max_foxes})")
        print(f"üìà Popula√ß√£o total final: {final_rabbits + final_foxes}")
        print(f"üìÅ Gr√°ficos salvos em: {os.path.abspath(self.image_dir)}")
        print("="*60)

        # Cria gr√°fico final resumo
        self.create_final_summary_plot()

        return {
            'final_step': final_step,
            'final_rabbits': final_rabbits,
            'final_foxes': final_foxes,
            'max_rabbits': max_rabbits,
            'max_foxes': max_foxes,
            'history': self.population_history.copy(),
            'image_directory': os.path.abspath(self.image_dir)
        }

    def create_final_summary_plot(self):
        """Cria gr√°fico final resumo da simula√ß√£o."""
        try:
            fig, ax = plt.subplots(1, 1, figsize=(14, 8))
            fig.patch.set_facecolor('white')

            if len(self.population_history['steps']) > 1:
                ax.plot(self.population_history['steps'],
                       self.population_history['rabbits'],
                       'g-', label='üê∞ Coelhos', linewidth=3, alpha=0.8)
                ax.plot(self.population_history['steps'],
                       self.population_history['foxes'],
                       'r-', label='ü¶ä Raposas', linewidth=3, alpha=0.8)

                # Marca pontos de m√°ximo
                max_rabbits = max(self.population_history['rabbits'])
                max_foxes = max(self.population_history['foxes'])
                max_r_step = self.population_history['steps'][self.population_history['rabbits'].index(max_rabbits)]
                max_f_step = self.population_history['steps'][self.population_history['foxes'].index(max_foxes)]

                ax.plot(max_r_step, max_rabbits, 'go', markersize=12,
                       label=f'Max Coelhos: {max_rabbits} (Step {max_r_step})')
                ax.plot(max_f_step, max_foxes, 'ro', markersize=12,
                       label=f'Max Raposas: {max_foxes} (Step {max_f_step})')

            ax.set_title("üê∞ü¶ä Simula√ß√£o Predador-Presa - Resultado Final",
                        fontsize=16, fontweight='bold')
            ax.set_xlabel("Steps", fontsize=12)
            ax.set_ylabel("Popula√ß√£o", fontsize=12)
            ax.legend(fontsize=11)
            ax.grid(True, alpha=0.3)

            plt.tight_layout()

            # Salva gr√°fico final
            final_filename = f"{self.image_dir}/FINAL_simulation_summary.png"
            plt.savefig(final_filename, dpi=300, bbox_inches='tight', facecolor='white')

            print(f"üìä Resumo final salvo: {final_filename}")

            # Tenta exibir
            try:
                display(Image(final_filename))
                plt.show()
            except:
                print("üì± Gr√°fico salvo - visualize o arquivo para ver o resultado final")

            plt.close(fig)

        except Exception as e:
            print(f"‚ùå Erro ao criar gr√°fico final: {e}")

    def list_saved_images(self):
        """Lista todas as imagens salvas da simula√ß√£o."""
        try:
            files = [f for f in os.listdir(self.image_dir) if f.endswith('.png')]
            files.sort()

            print(f"\nüìÅ Imagens salvas em {os.path.abspath(self.image_dir)}:")
            for i, file in enumerate(files, 1):
                print(f"  {i:3d}. {file}")

            print(f"\nüìä Total: {len(files)} imagens")

        except Exception as e:
            print(f"‚ùå Erro ao listar imagens: {e}")

    def clean_saved_images(self):
        """Remove todas as imagens salvas da simula√ß√£o."""
        try:
            files = [f for f in os.listdir(self.image_dir) if f.endswith('.png')]

            for file in files:
                os.remove(os.path.join(self.image_dir, file))

            print(f"üßπ Removidas {len(files)} imagens do diret√≥rio {self.image_dir}")

        except Exception as e:
            print(f"‚ùå Erro ao limpar imagens: {e}")
```

```python
#%%writefile simulator.py
from field import Field


# simulator.py
class Simulator:
    """
    Um simulador simples predador-presa, baseado em um campo retangular
    contendo coelhos e raposas.
    """

    # Constantes representando informa√ß√µes de configura√ß√£o para a simula√ß√£o
    DEFAULT_WIDTH = 120   # A largura padr√£o da grade
    DEFAULT_DEPTH = 80    # A profundidade padr√£o da grade
    FOX_CREATION_PROBABILITY = 0.02      # A probabilidade de uma raposa ser criada
    RABBIT_CREATION_PROBABILITY = 0.08   # A probabilidade de um coelho ser criado

    def __init__(self, depth=None, width=None):
        """
        Cria um campo de simula√ß√£o com o tamanho dado.
        """
        if depth is None:
            depth = self.DEFAULT_DEPTH
        if width is None:
            width = self.DEFAULT_WIDTH

        if width <= 0 or depth <= 0:
            print("As dimens√µes devem ser >= zero.")
            print("Usando valores padr√£o.")
            depth = self.DEFAULT_DEPTH
            width = self.DEFAULT_WIDTH

        self.field = Field(depth, width)
        self.view = SimulatorView(depth, width)
        self.step = 0

        self.reset()

    def run_long_simulation(self):
        """
        Executa a simula√ß√£o a partir de seu estado atual por um
        per√≠odo razoavelmente longo (4000 passos).
        """
        self.simulate(4000)

    def simulate(self, num_steps):
        """
        Executa a simula√ß√£o pelo n√∫mero dado de passos.
        Para antes do n√∫mero dado de passos se deixar de ser vi√°vel.
        """
        self.report_stats()
        for n in range(1, num_steps + 1):
            if not self.field.is_viable():
                break
            self.simulate_one_step()
            self._delay(100)  # ajuste isto para mudar a velocidade de execu√ß√£o

    def simulate_one_step(self):
        """
        Executa a simula√ß√£o a partir de seu estado atual por um √∫nico passo.
        Itera sobre todo o campo atualizando o estado de cada raposa e coelho.
        """
        self.step += 1
        # Usa um Field separado para armazenar o estado inicial do pr√≥ximo passo
        next_field_state = Field(self.field.get_depth(), self.field.get_width())

        rabbits = self.field.get_rabbits()
        foxes = self.field.get_foxes()

        # Deixa todos os coelhos correrem
        for rabbit in rabbits:
            rabbit.run(self.field, next_field_state)

        # Deixa todas as raposas ca√ßarem
        for fox in foxes:
            fox.hunt(self.field, next_field_state)

        # Substitui o estado antigo pelo novo
        self.field = next_field_state

        self.report_stats()
        self.view.show_status(self.step, self.field)

    def reset(self):
        """Redefine a simula√ß√£o para uma posi√ß√£o inicial."""
        self.step = 0
        self._populate()
        self.view.show_status(self.step, self.field)

    def report_stats(self):
        """Relata o n√∫mero de cada tipo de animal no campo."""
        print(f"Step: {self.step} ", end="")
        self.field.field_stats()

    def _populate(self):
        """Popula aleatoriamente o campo com raposas e coelhos."""
        rand = Randomizer.get_random()
        self.field.clear()

        for row in range(self.field.get_depth()):
            for col in range(self.field.get_width()):
                if rand.random() <= self.FOX_CREATION_PROBABILITY:
                    location = Location(row, col)
                    fox = Fox(True, location)
                    self.field.place_fox(fox, location)
                elif rand.random() <= self.RABBIT_CREATION_PROBABILITY:
                    location = Location(row, col)
                    rabbit = Rabbit(True, location)
                    self.field.place_rabbit(rabbit, location)
                # sen√£o deixa a localiza√ß√£o vazia

    def _delay(self, milliseconds):
        """
        Pausa por um tempo dado.
        :param milliseconds: O tempo para pausar, em milissegundos
        """
        time.sleep(milliseconds / 1000.0)


# main.py - exemplo de uso
if __name__ == "__main__":
    print("=== Simula√ß√£o Foxes and Rabbits ===")
    print("Comandos dispon√≠veis:")
    print("1 - Simular um passo")
    print("2 - Simular N passos")
    print("3 - Executar simula√ß√£o longa (4000 passos)")
    print("4 - Resetar simula√ß√£o")
    print("0 - Sair")

    # Cria o simulador
    simulator = Simulator(40, 60)  # Campo menor para melhor visualiza√ß√£o

    while True:
        try:
            choice = input("\nEscolha uma op√ß√£o: ").strip()
            if choice == "0":
                print("Encerrando simula√ß√£o...")
                break
            elif choice == "1":
                simulator.simulate_one_step()
            elif choice == "2":
                try:
                    steps = int(input("N√∫mero de passos: "))
                    if steps > 0:
                        simulator.simulate(steps)
                    else:
                        print("N√∫mero de passos deve ser positivo.")
                except ValueError:
                    print("Por favor, digite um n√∫mero v√°lido.")
            elif choice == "3":
                print("Iniciando simula√ß√£o longa...")
                simulator.run_long_simulation()
            elif choice == "4":
                simulator.reset()
                print("Simula√ß√£o resetada.")
            else:
                print("Op√ß√£o inv√°lida. Tente novamente.")

        except KeyboardInterrupt:
            print("\n\nSimula√ß√£o interrompida pelo usu√°rio.")
            break
        except Exception as e:
            print(f"Erro: {e}")

    print("Simula√ß√£o finalizada.")
```
